```python
import face_recognition as face_rec
import cv2
import matplotlib.pyplot as plt
```

```python
image = face_rec.load_image_file("rsc/img01.jpeg")
```

```python
face_locations = face_rec.face_locations(image)
```

```python
face_locations
```

```python
top,right,bottom,left = face_locations[0]
```

```python
cv2.rectangle(image, (left,top), (right,bottom), (30,100,255),2)
cv2.rectangle(image, (left,bottom-35), (right,bottom), (30,100,255), cv2.FILLED)
font = cv2.FONT_HERSHEY_TRIPLEX
cv2.putText(image, 'face', (left+10, bottom-5), font, 1.0, (255,255,255),1)
```

```python
plt.imshow(image)
```

```python
cv2.imwrite('obama.png',image)
```
